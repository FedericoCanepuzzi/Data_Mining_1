```python
df_cluster=df[["Satisfaction"]]
scaler = RobustScaler()
X = scaler.fit_transform(df_cluster)
data_dist = pdist(X, metric='euclidean')
data_link = linkage(data_dist, method='complete', metric='euclidean')
res = dendrogram(data_link, color_threshold=2, truncate_mode='lastp')
```

```python
color_threshold = 1.3
num_clusters = 5
clusters = fcluster(data_link, color_threshold, criterion='distance')
#df['VehBCost-Class'] = clusters
df['Satisfaction-Class'] = clusters

mapClassName = {}
for i in range(1, num_clusters+1):
    #classVehBCost = df[df['VehBCost-Class'] == i]['VehBCost']
    classSatisfaction=df[df['Satisfaction-Class'] == i]['Satisfaction']
    
    mapClassName[i] = "Satisfaction[" + str(classSatisfaction.min()) + "; " + str(classSatisfaction.max()) + "]"
df['Satisfaction-Class'] = df['Satisfaction-Class'].map(mapClassName).astype(str)
```
